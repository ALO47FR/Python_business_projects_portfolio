```python
# La colonne Valeur contient la valeur semi-quantitative <0.1

display(sousNut.loc[1085:1095, :])
```

```python
# Besoin de la colonne Valeur en numérique pour calculer la proportion de sous-nutrition

# Remplacement arbitraire de la valeur semi-quantitative <0.1 par 0, la quantité étant considérée comme négligeable.

sousNut.Valeur[sousNut.Valeur == '<0.1'] = '0.0'
sousNut['Valeur'] = sousNut['Valeur'].apply(pd.to_numeric)
display(sousNut.loc[1085:1095, :])
```

La colonne 'Valeur' des cases vides quand il n'y a pas de sous-nutrition ou qu'il n'y a pas de données.  
Même si il est impossible de discriminer avec certitude la valeur 0 d'une valeur 'absence de données', ces deux types de données sont considérées comme négligeables. Cette colonne sera  modifiée en remplaçant les valeurs manquantes par 0.

```python
# Remplacement des données manquantes par des zéros 

sousNut = sousNut.fillna(0.0)
display(sousNut.loc[1085:1095, :])
```

   <span style='color:chocolate; font-family:Microsoft JhengHei UI Light; font-size:16px '> <ins><I>Problème 3</I></ins>: L'année n'est pas une valeur numérique </span>

```python
# Renommer la colonne 'Année' en Range 
# Décomposer le Range en deux colonnes

sousNut.rename(columns={'Année':'Range des Années'}, inplace=True)
split = sousNut['Range des Années'].str.split('-', expand=True)
split.insert(2,'minAn',split[0],allow_duplicates=False)
split.insert(3,'maxAn',split[1],allow_duplicates=False)
split[['minAn','maxAn']] = split[['minAn','maxAn']].apply(pd.to_numeric)
split = split.loc[:,['minAn','maxAn']]
print(split)
```

```python
# Faire la moyenne des bornes du range et transformer en entier

moyAn = split.mean(axis=1)
moyAn = moyAn.astype('int32')
print(moyAn)
```

```python
# Insérer la nouvelle colonne dans le tableau de sous-nutrition

sousNut.insert(1,'Année',moyAn,allow_duplicates=False)
sousNut.head(5)
```

```python
# Vérification de la transformation

sousNut.dtypes
```

## <a name="A221"><span style='color:darkdodgerblue;font-family:Microsoft JhengHei UI Light'> II. 2. 1. Equations régissant la disponibilité intérieure</span></a>

La disponibilité intérieure correspond à la balance entre les entrées et les sorties. 

<span style='color:chocolate; font-family:Microsoft JhengHei UI Light; font-size:14px '> Disponibilité intérieure = Production + Importations - Exportations + Variation du stock</span>

La disponibilité intérieure est destinée à différents usages et les quantités consommées ou perdues correspondent donc aux quantités qui étaient initialement disponibles.

<span style='color:chocolate; font-family:Microsoft JhengHei UI Light; font-size:14px '> Disponibilité intérieure = Nourriture + Alimentation animaux + Semences + Traitement (transformation en d'autres produits) + Autres utilisations + Pertes </span>

```python
#Assemblage des données de population et d'insécurité alimentaire

mergePopSN = pd.merge(pop, sousNut, how='outer', on=['Zone','Année'])
mergePopSN = mergePopSN.drop(['Range des Années'], axis=1)
mergePopSN = mergePopSN.dropna()
```

```python
#Sélection de 2017

popSN2017 = mergePopSN.query('Année == 2017.0')

#Calcul de la proportion sur la totalité des données

popTot2017 = popSN2017['Population (en milliers habitants)'].sum().round(3)
SNTot2017 = popSN2017['Personnes sous-alimentées (en milliers habitants)'].sum().round(3)
propTotSN2017 = (100*SNTot2017/popTot2017).round(2)

print('\nEn 2017, au moins', propTotSN2017, '% de le population mondiale est en état de sous-nutrition.') 

dic = {'Année':'2017','Population mondiale (en milliers habitants)': popTot2017,'Population en sous-nutrition (en milliers habitants)': SNTot2017,'Proportion de personnes en état de sous-nutrition':(propTotSN2017).astype(str) +' %'}
display(pd.DataFrame([dic]))
```

## <a name="A312"><span style='color:darkdodgerblue;font-family:Microsoft JhengHei UI Light'> III. 1. 2. Théoriquement, il ne devrait pas y avoir de personne en état de sous-nutrition </span></a>

<b><i>N.B.</i> </b> Les besoins énergétiques alimentaires moyens sont établis par la FAO dépendent, entre autres facteurs, du pays considéré. Une approximation les estiment toutefois à l'échelle mondiale à 50g/jour de protéines (FAO, 2013). Les besoins caloriques moyens sont plus hétérogènes mais nous prendrons une valeur moyenne arbitraire de 2500 kcal par jour dans le cadre de ce rapport.

<span style='color:grey;font-size:9px'> FAO, FIDA et PAM. 2013. L’état de l’insécurité alimentaire dans le monde 2013. Les multiples dimensions de la sécurité alimentaire. Rome. FAO. E-ISBN 978-92-5-207917-0 (PDF)</span>

### *Vue la disponibilité alimentaire mondiale, 8.3 milliards de personnes (111% de la population mondiale) pourraient être nourries ...*

```python
#Assemblage des données de population et de disponibilité

pop2017a = pop.query('Année == 2017.0')
pop2017a = pop2017a.groupby(['Zone']).mean()
dispo2017a = dispo.fillna(0)
dispo2017a = dispo2017a.groupby(['Zone']).sum()

mergePopDispo = pd.merge(pop2017a,dispo2017a, how='outer', on=['Zone'])
mergePopDispo = mergePopDispo.dropna(subset=['Année'])
mergePopDispo = mergePopDispo.sort_values('Disponibilité alimentaire (Kcal/personne/jour)', ascending=True)
DP2017 = mergePopDispo.fillna(0)

# Calcul des calories totales disponibles par pays

pop2017b = 1000*DP2017['Population (en milliers habitants)']
cal2017a = DP2017['Disponibilité alimentaire (Kcal/personne/jour)']
cal2017b = cal2017a*pop2017b

# Calcul des calories disponibles à l'échelle mondiale par habitant
calTot2017 = cal2017b.sum()
popTot2017 = pop2017b.sum()

DispoAlimTot2017 = (calTot2017/popTot2017).astype(int)

print('\nEn 2017, les ressources alimentaires mondiales disponibles auraient permis à chaque habitant de manger', DispoAlimTot2017,'Kcal par jour.')
```

```python
# Sachant que les besoins minimaux moyens sont de 1819kcal/jour/personne

dispoMin = 2500
persTheo = (calTot2017/dispoMin)
persTheo = persTheo.astype('int64')
propAlim = (persTheo/popTot2017)
propAlim = propAlim.round(2)
print("")
print('Autrement dit,', persTheo,'personnes pourraient théoriquement être nourries avec la disponibilité alimentaire présente à une \néchelle mondiale; ce qui, en 2017, représente environ', propAlim, 'fois la population mondiale.')
```

### *Cas particulier de la disponibilité intérieure en céréales*

```python
# Création d'un dataframe avec les données céréales

dfDispoCereales = dispo.query('Produit == ["Avoine","Blé","Orge","Maïs","Millet","Riz (Eq Blanchi)","Seigle","Sorgho","Céréales, Autres"]')
dfDispoCereales = dfDispoCereales.loc[:,['Zone', 'Produit', 'Disponibilité intérieure','Nourriture (en milliers de tonnes)', 'Aliments pour animaux (en milliers de tonnes)']]



# Utilisations en alimentations humaine et animale, par produits céréaliers, au niveau mondial

dfDispoCerealesTot = (dfDispoCereales.groupby(dfDispoCereales['Produit']).sum()).astype(int)
dfDispoCerealesTot = dfDispoCerealesTot.reset_index()

propCerealesNour = (100*dfDispoCerealesTot['Nourriture (en milliers de tonnes)']/dfDispoCerealesTot['Disponibilité intérieure']).astype(int)
dfDispoCerealesTot.insert(3,"Part destinée à l'alimentation humaine (en %)",propCerealesNour,allow_duplicates=False)

propCerealesAnimx = (100*dfDispoCerealesTot['Aliments pour animaux (en milliers de tonnes)']/dfDispoCerealesTot['Disponibilité intérieure']).astype(int)
dfDispoCerealesTot.insert(5,"Part destinée à l'alimentation animale (en %)",propCerealesAnimx,allow_duplicates=False)

# Calcul de la part disponible dédiée aux alimentations animale et humaine, tous types de céréales confondus

cerealesDispoInter = dfDispoCerealesTot['Disponibilité intérieure'].sum()
dispoCerealesNourTot = dfDispoCerealesTot['Nourriture (en milliers de tonnes)'].sum()
dispoCerealesAnimxTot = dfDispoCerealesTot['Aliments pour animaux (en milliers de tonnes)'].sum()

propCerealesNourTot = (100*dispoCerealesNourTot/cerealesDispoInter).round(1)
propCerealesAnimxTot = (100*dispoCerealesAnimxTot/cerealesDispoInter).round(1)

print("\nAu niveau mondial, les produits céréaliers destinés à l'alimentation se répartissent de manière quasi-égale entre alimentation \nhumaine (",propCerealesNourTot," %) et animale (",propCerealesAnimxTot ," %).")
print("La répartition par produit céréalier est indiquée par produit dans le tableau ci-dessous:")
display(dfDispoCerealesTot)
```

### *Cas particulier du manioc en Thaïlande en 2017 : une ressource alimentaire importante exportée*

```python
# Création d'un dataframe avec les données du manioc et des populations, en Thaïlande 
popThailande2017 = pop.query('Zone == ["Thaïlande"] and Année == 2017')
dispoManiocThailande = dispo.query('Produit == ["Manioc"] and Zone == ["Thaïlande"]')
sousNutThailande = sousNut.query('Zone == ["Thaïlande"] and Année ==2017')

MT2017 = pd.merge(dispoManiocThailande, sousNutThailande, how='outer', on=['Zone'])
MT2017 = MT2017.reset_index()
MTP2017 = pd.merge(MT2017,popThailande2017, how='outer', on=['Zone'])
MTP2017 = MTP2017.reset_index()
MTP2017 = MTP2017.drop(['level_0', 'index'], axis=1)

#Liste de variable

MTNour = MTP2017['Nourriture (en milliers de tonnes)'].sum()
MTAnimx = MTP2017['Aliments pour animaux (en milliers de tonnes)'].sum()
MTTot = MTP2017['Disponibilité intérieure'].sum()
MTExport = MTP2017['Exportations (en milliers de tonnes)'].sum()
MTProd = MTP2017['Production (en milliers de tonnes)'].sum()
MTDispoAlim = MTP2017['Disponibilité alimentaire (Kcal/personne/jour)'].sum()
MTPop = MTP2017['Population (en milliers habitants)'].sum()
MTsousNut = MTP2017['Personnes sous-alimentées (en milliers habitants)'].sum()


propMTExport = MTExport/MTTot
propMTExport = propMTExport.astype(int)
propMTExpProd = 100*MTExport/MTProd
propMTExpProd = propMTExpProd.astype(int)

MTExportInt = (MTExport/1000).astype(int)

# 871 milliers de tonnes de nourriture donne 13 Kcal/personne/jour
# Combien donnent les 25 214 milliers de tonnes exportées si elles sont utilisés pour l'alimentation humaine?

MTExportCal = MTExport*MTDispoAlim/MTNour
MTExportCal = MTExportCal.round(0)

# Sachant que les besoins minimaux moyens sont de 2500kcal/jour/personne
# On cherche combien de personnes ont peut nourrir

dispoMin = 2500
persTheoMT = (MTExportCal*MTPop)/dispoMin
persTheoMT = persTheoMT.astype(int)

propMTpers = (persTheoMT/MTsousNut).astype(int)

propPopMT2017 = (100*persTheoMT/MTPop).round(1)



print("\nIl y a",propMTExport,"fois plus de manioc exporté que de manioc utilisé sur place.")
print("C'est à dire que plus de", propMTExpProd,"% de la production de manioc Thaïlandaise est exportée.")
print("Pourtant, en 2017,",(100*MTsousNut/MTPop).round(2),"% de la population Thaïlandaise est en état de sous-nutition.")
print("\nSi on utilisait les",MTExportInt,"millions de tonnes de manioc exportées pour nourrir les thaïlandais, leur apport calorique augmenterait de", MTExportCal,"Kcal/personne/jour.")
print("\nEn 2017, ces quelques",MTExportInt,"millions de tonnes exportées auraient pu ainsi nourrir",persTheoMT,"thaïlandais. \nCela représente", propMTpers,"fois le nombre de personnes considérées comme sous-alimentées cette année-là \nou encore,", propPopMT2017, "% de la population thaïlandaise totale.")

display(MTP2017)
```
