### <a href="#RM1" name="M1" style='text-decoration:none'><span style='color:black'>*Test d'unicité*</span></a>

```python
# Effectuer un test d'unicité sur colonne clé candidate "cle" en vue de valider son utilisation en tant que clé primaire.
    
def test_unicite(dataframe, cle):
    
    doublons = dataframe[cle].duplicated().sum()
    
        # Afficher le résultat du test de manière explicite.

    if doublons > 0 :
            print("La colonne contient",doublons,"doublons. \nle critère d'unicité, obligatoire pour une clé primaire, n'est pas respecté.")          
           
    else:
            print("Le critère d'unicité est respecté. La clé primaire ne contient pas de doublons.")
          
```

### <a href="#RM2" name ="M2" style='text-decoration:none'><span style='color:black'>*Test pour définir la distribution la plus adéquate pour modéliser une variable*</span></a>

```python
def test_distribution(df, col, bins):
    
    # Définir les méthodes de validation à tester
    sum_square_error=[]
    p_value=[]
    
    # Définir une liste de loi à tester.
    dist_names = ['lognorm', 'norm', 'pareto', 't','invgauss', 'alpha','gamma','loggamma','invgamma', 'chi', 'chi2']  
    y, x= np.histogram(df[col], bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0
        
    # Tester les lois
    for distribution in dist_names:
                       
    # méthode fit 
        dist = getattr(scipy.stats, distribution)
        parameters = dist.fit(df[col])
               
    # paramètres
        loc = parameters[-2]
        scale = parameters[-1]
        arg = parameters[:-2]
                
    #Sum square error
        pdf = dist.pdf(x,*arg, loc=loc, scale=scale)
        sse = np.sum((y - pdf)**2)
        sum_square_error.append(sse)
            
    #p_value
        p=scipy.stats.kstest(df[col],distribution,parameters, N=bins)[1]
        p_value.append(p)
                       
     # Supprimer le warning de performance runtime                
        import warnings
        warnings.filterwarnings("ignore")
        
    dfResults = pd.DataFrame()
    
    # Vectorization
    dfResults = pd.DataFrame()
    dfResults['Distribution'] = dist_names
    dfResults['Erreur quadratique moyenne'] = sum_square_error
    dfResults['p_value'] = p_value
       
    dfResults.sort_values('Erreur quadratique moyenne', inplace=True, ascending=True)
    dfResults = dfResults.reset_index()
    dfResults = dfResults.drop('index', axis=1)

    # rapport
    print ('Classement des distributions selon leur pertinence pour modéliser la distribution de',col,':')
    display(dfResults)
    if (((dfResults['Erreur quadratique moyenne'].iloc[0]<0.05) & (dfResults['p_value'].iloc[0]> 0.05)) & (dfResults['Erreur quadratique moyenne'].iloc[1]<0.05) & (dfResults['p_value'].iloc[1]> 0.05)):
        print("Les lois",dfResults['Distribution'].iloc[0],"et",dfResults['Distribution'].iloc[1],"sont les plus proches de la distribution testée.")
    
    elif ((dfResults['Erreur quadratique moyenne'].iloc[0]<0.05) & (dfResults['p_value'].iloc[0]> 0.05)): 
        print("La loi",dfResults['Distribution'].iloc[0],"est la plus proche de la distribution testée.") 
    
    else:
        print("Aucune des distributions testées ne semble pertinente.\nLe mélange de gaussienne reste une hypothèse à tester.")
 
  
     
   
```

### <a href="#RM3" name ="M3" style='text-decoration:none'><span style='color:black'>*Sélectionner les colonnes d'un tableau dans une popup et afficher le tableau customisé*</span></a>

```python
def Format_table_sortie(df):

    #Format de la fenêtre 

    window=Tk()
    window.title("Caractéristiques")
    list_box = Listbox(window, selectmode = 'multiple',height = 10, yscrollcommand=True)
    labelChoix = Label(window, text = "Veuillez choisir quelles caractéristiques vous voulez afficher pour décrire les outliers:")
    labelChoix.pack()


    #Actions

    def submit_and_close():
        select = []
        for colonne in list_box.curselection():
            select.insert(colonne,list_box.get(colonne))
        print("Afin d'afficher un tableau customisé des résultats, vous avez sélectionné (dans la fenêtre popup tkinter) les colonnes :")
        for colonne in select:
            print(colonne)
        sortie = df[select].reset_index()
        sortie = sortie.drop('index', axis=1)
        display(sortie)
        window.destroy()   

    def cancel():
        window.destroy()

    def selectAll():
        list_box.select_set(0, tk.END)
        
    def deselectAll():
        list_box.selection_clear(0, tk.END)

# Contenu de la fenêtre
    #Menu déroulant
    column_list = df.columns.tolist()
    n=len(column_list)
    for items in (column_list):
        list_box.insert(END,items)
    list_box.pack()

        # Bouton de valider et quitter
    submit_button = Button(window,text="Valider la sélection", command=submit_and_close)
    submit_button.pack()

        # Bouton d'annulation
    cancel_button = Button(window,text="Annuler et fermer",command=cancel)
    cancel_button.pack()   

        # Bouton selectionner tout
    selecAll_button = Button(window,text="Sélectionner tout", command = selectAll)
    selecAll_button.pack()

        # Bouton déselectionner tout
    deselecAll_button = Button(window,text="Désélectionner tout", command = deselectAll)
    deselecAll_button.pack()    
      
    window.mainloop()
```

### <a href="#RM4" name ="M4" style='text-decoration:none'><span style='color:black'> *Détection des valeurs abérrantes via les Z-scores et renvoi sous forme d'un tableau customisé*</span></a>

```python
def detect_outliers_Z(df, col):
    
    dfZ = df
    mean = np.mean(dfZ[col])
    std = np.std(dfZ[col])
    
#N.B. test standard pour où le threshold = 3 
#N.B. impact sur l'analyse si la distribution réelle n'est pas symétrique ou ne suit pas une loi normale
    
    dfZ['z_score'] = (dfZ[col]-mean)/std
    
    df_outliersZ = dfZ.query('z_score >= 3').sort_values(col)
    
    print("\nPar la méthode des Z-scores,",df_outliersZ['z_score'].count(),"outliers sont détectés. \nCi-dessous, le détail des lignes outliers sur la variable",col,".\n")
    
    # Visualisation des outliers dans un tableau
    Format_table_sortie(df_outliersZ)
    
    # Nouvelle distribution sans les outliers
    print('La nouvelle distribution de la variable',col,'sans ces outliers (Z-score>3) est:')
    df_Z_dist = dfZ.loc[(dfZ['z_score']<3),:]
    plt.hist(dfZ[col],bins=100,rwidth=0.8)
    plt.title("Distribution sans les outliers",fontsize=20)
    plt.xlabel(col)
    plt.ylabel("Effectif")
    plt.show()
    test_distribution(df_Z_dist, 'price', 500)
      
```

### <a href="#RM5" name ="M5" style='text-decoration:none'><span style='color:black'>*Détection des valeurs abérrantes via la méthode IQR et renvoi sous forme d'un tableau customisé*</span></a>

```python
def detect_outliers_IQR(df, col):
    
    #Définir la médiane et la moyenne
    Median = df[col].quantile(0.5).round(2)
    Moyenne = np.around(np.mean(df[col]), decimals=2)
    Ecart = np.around(np.abs(Moyenne-Median), decimals=2)
    print("La médiane est", Median, "et la moyenne est:",Moyenne,". L'écart observé (en valeur absolue) est de",Ecart,".")

    #Définir le 1er quantile ou 25eme percentile
    Q1 = df[col].quantile(0.25)
    print("Le 25eme percentile est", Q1.round(2))
    
    #Définir le 3eme quantile ou 75eme percentile
    Q3 = df[col].quantile(0.75)
    print("Le 75eme percentile est", Q3.round(2))
    
    #Définir l'intervalle InterQuartile
    IQR = Q3-Q1
    print("L'Intervalle InterQuartile est:",IQR.round(2))

    # Visualisation des outliers dans un tableau
    
    df_outliersIQR = df.loc[(df[col] >= Q3 + 1.5*IQR)|(df[col] <= Q1 - 1.5*IQR),:]
    
    nb_outliersIQR = df_outliersIQR[col].count()
    
    print("\nAprès analyse de la caractéristique",col,",",nb_outliersIQR,"outliers potentiels sont identifiés et présentés ci-dessous.\n")

    df_outliersIQR = df_outliersIQR.sort_values(col)
    
    # Visualisation des outliers dans un tableau
    Format_table_sortie(df_outliersIQR)
       
    # Nouvelle distribution sans les outliers avec modèle
    print('La nouvelle distribution de la variable',col,'sans ces outliers est: ')
    df_IQR_dist = df.loc[(df[col] <= Q3 + 1.5*IQR)&(df[col] >= Q1 - 1.5*IQR),:]
    plt.hist(df_IQR_dist[col],bins=100,rwidth=0.8)
    plt.title("Distribution sans les outliers",fontsize=20)
    plt.xlabel(col)
    plt.ylabel("Effectif")
    plt.show()
    test_distribution(df_IQR_dist, 'price', 500)
       
```

#### *Détermination de la loi statistique pouvant modéliser les données relatives au prix*

<a href="#M2" name="RM2"> Voir la méthode </a>

```python
test_distribution(dfWebDfErp, 'price', 500)
```

```python
# Représentation graphique
bins=500
sse = np.inf
sse_thr = 0.010

# Pour chaque distribution
data = dfWebDfErp['price']
dist_names = ['lognorm','invgauss']

### Histogramme des données
y, x = np.histogram(data, bins=bins, density=True)
# Milieu de chaque classe
x = (x + np.roll(x, -1))[:-1] / 2.0


for name in dist_names:

    # Modéliser
    dist = getattr(scipy.stats, name)
    param = dist.fit(data)

    # Paramètres
    loc = param[-2]
    scale = param[-1]
    arg = param[:-2]

    # PDF
    pdf = dist.pdf(x, *arg, loc=loc, scale=scale)
    # SSE
    model_sse = np.sum((y - pdf)**2)

    # Si le SSE est ddiminué, enregistrer la loi
    if model_sse < sse :
        best_pdf = pdf
        sse = model_sse
        best_loc = loc
        best_scale = scale
        best_arg = arg
        best_name = name

    # Si en dessous du seuil, quitter la boucle
    if model_sse < sse_thr :
        break
        
plt.figure(figsize=(12,8))
plt.plot(x, y, label="Données")
plt.plot(x, best_pdf, label=best_name, linewidth=3)
plt.legend(loc='upper right')
plt.title("Représentation des données et du modèle créé selon la loi inverse gaussienne", fontsize=20)
plt.xlabel("Prix des produits (€)")
plt.ylabel("Nombre de produits")
plt.show()

# Détails sur la loi sélectionnée
print("Selected Model : ", best_name)
print("Loc. param. : ", best_loc)
print("Scale param. : ", best_scale)
print("Other arguments : ", best_arg)
print("SSE : ", sse)
```

<span style = 'color:chocolate; font-family:Microsoft JhengHei UI Light'> <b> ► Conclusion </b>: La distribution ne suit pas une loi normale mais une loi de type "à queue lourde" pouvant correspondre à plusieurs modèles de distribution, en particulier inverse-gaussienne et lognormale. </span>
