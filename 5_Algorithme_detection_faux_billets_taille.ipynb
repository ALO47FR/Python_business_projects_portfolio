## Contexte du projet
L’Organisation nationale de lutte contre le faux-monnayage (ONCFM) vise, en tant qu'organisation publique, de mettre en place des méthodes d’identification des contrefaçons des billets en euros.  


## Objectifs
Les dimensions de ces contrefaçons sont souvent légèrement différentes des billets authentiques. 

Notre objectif est donc de créer un algorithme qui soit capable de différencier automatiquement les vrais des faux billets à partir de leurs dimensions géométriques.  

Cet algorithme se devra d’être naturellement le plus performant possible
pour identifier un maximum de faux billets au sein de la masse de billets
analysés chaque jour.  


## Démarche
Lorsqu’un billet arrive, une machine mesure avec précision l’ensemble
de ses caractéristiques géométriques. 

Ces données sont consignées dans un fichier et seront analysées par l'algorithme créé.  

Lors de cette première étape de création deux types d'algorythme seront testés :   

● une régression logistique classique, qui est une méthode d'apprentissage supervisé ;  

● un k-means, qui est une méthode d'apprentissage non supervisé et duquel seront utilisés les centroïdes pour réaliser la
prédiction.    

La paramétrisation sera faite sur un jeu de données de 1500 billets (1000 vrais et 500 faux). 

Pour une évaluation optimale des modèles, une analyse des nombres de faux positifs et faux négatifs sera réalisée *via* une matrice de confusion.

### Régression linéaire optimale stepwise (type backward)

```python
def backward_VarSelectionRLM(df, y):
    
    # Modèle de régression multiple stepwise optimisé par méthode backward
        # On commence par la régression avec tous les paramètres et on ote pas à pas les variables les moims significatives (p-value la plus forte)
        
       
    Xs = set(df._get_numeric_data().columns)
    if y in Xs:
        Xs.remove(y)
    cond = True

    while Xs and cond:
        formule = "{} ~ {} + 1".format(y,' + '.join(Xs))
        print('\nModèle de régression multiple testé : ')
        print(formule)
        model_multi = smf.ols(formule, df).fit()
        score = model_multi.pvalues[1:]
        Enlever = score[score == score.max()]
        if Enlever.values > 0.01:
            print('La variable', Enlever.index[0], "n'est pas significative (p-value :", round(Enlever.values[0],3),")")
            Xs.remove(Enlever.index[0])
        else:
            cond = False
            print('est le modèle de régression linéaire multiple optimisé par la méthode stepwise en backward.')
        print('')
    print(model_multi.summary())
    
    return model_multi
```

### Conditions de validités d'un modèle de regression linéaire (simple ou multiple)

```python
def ValiditeRL(reg, nb_Variables):
    
    print("\nVérifications des conditions de validité d'une régression linéaire :")
    
    # Objets :
    
    cond1 = "\nCondition 1 : La relation entre les deux variables numériques continues est réellement de forme linéaire.\n"
    cond2 = "\nCondition 2 : Les résidus de la régression linéaire sont indépendants.\n"
    cond3 = "\nCondition 3 : Les résidus de la régression linéaire sont distribués selon une loi Normale d'espérance nulle.\n"
    cond4 = "\nCondition 4 : Les résidus de la régression linéaire sont distribués de façon homogène (homoscédasticité).\n"
    cond5 = "\nCondition 5 : Il n'y a pas de collinéarité entre les variables explicatives."
    
    dfValidite = pd.DataFrame()
    Validite_list = []
    
    
    #Test 1
    if nb_Variables == 1:
        print(cond1)
        print("Test avec le multiplicateur de Harvey-Collier :\nH0 : La condition de linéarité est vérifiée (p>0.05) \nH1 : La condition de linéarité n'est pas vérifiée.")
        name = ["t value", "p value"]
        testHC = sms.linear_harvey_collier(reg)
        print('La p-value issue du test de Harvey-Collier est:',testHC[1])
        if testHC[1] < 0.05 :
            print("H0 est rejetée, la condition de linéarité n'est pas vérifiée. Une approche graphique est à réaliser.")
            Validite_list.append({'Condition':'Linéarité', 'Validité': 'A vérifier'})
        else: 
            print("H0 est acceptée, la condition de linéarité est vérifiée.")
            Validite_list.append({'Condition':'Linéarité','Validité' : 'Oui'})
    else: 
        print(cond1)
        print("Test avec le multiplicateur de Lagrange :\nH0 : La condition de linéarité est vérifiée (p>0.05) \nH1 : La condition de linéarité n'est pas vérifiée.")
        name = ["t value", "p value"]
        testHC = sms.linear_lm(reg.resid, reg.model.exog)
        print('La p-value issue du test de Lagrange est:',testHC[1])
        if testHC[1] < 0.05 :
            print("H0 est rejetée, la condition de linéarité n'est pas vérifiée. Une approche graphique est à réaliser.")
            Validite_list.append({'Condition':'Linéarité', 'Validité': 'A vérifier'})
        else: 
            print("H0 est acceptée, la condition de linéarité est vérifiée.")
            Validite_list.append({'Condition':'Linéarité','Validité' : 'Oui'})
    
    
    # Test 2 
    print(cond2)
    print("Évaluation par la statistique de Durbin-Watson.")
    testDW = sms.durbin_watson(reg.resid, axis=0)
    print("Statistique de Durbin-Watson :",testDW)
    print("La statistique de Durbin-Watson équivaut à 2*(1-r) donc le coefficient de correlation r vaut approximativement:", round(1-(testDW/2),2))
    if ((testDW > 1.8)&(testDW < 2.2)) :
        print("Il existe peu de relaton entre les résidus (r < 0.1), la condition d'indépendance est vérifiée.")
        Validite_list.append({'Condition':'Indépendance des résidus' ,'Validité' : 'Oui'})
    elif ((testDW > 1.34)&(testDW < 2.66)) : 
        print("Il existe possiblement une relaton entre les résidus (r < 0.33), la condition d'indépendance n'est peut-être pas vérifiée. Une approche graphique est à réaliser.")
        Validite_list.append({'Condition':'Indépendance des résidus' ,'Validité' : 'A vérifier'})
    else : 
        print("Il existe une relaton entre les résidus (r > 0.33), la condition d'indépendance n'est pas vérifiée. Une approche graphique est à réaliser.")
        Validite_list.append({'Condition':'Indépendance des résidus' ,'Validité' : 'A vérifier'})
     
   
    # Test 3
    print(cond3)
    m = reg.resid.mean()
    print("La moyenne des résidus est :",m)
    if abs(m) < 0.0001 : 
        print("L'espérance des résidus est donc négligeable voire nulle.")
        Validite_list.append({'Condition':'Espérance nulle des résidus' ,'Validité' : 'Oui'})
    else : 
        print("L'espérence des résidus n'est pas nulle.\n")
        Validite_list.append({'Condition':'Espérance nulle des résidus','Validité' : 'A vérifier'})
    print("La condition de normalité est évaluée par le test de Jarque-Bera :\nH0 : La condition de normalité des résidus est vérifiée (p>0.05) \nH1 : La condition de normalité des résidus n'est pas vérifiée.")
    testJB = sms.jarque_bera(reg.resid)
    print('La p-value issue du test de Jarque-Bera est:',testJB[1])
    if testJB[1] < 0.05 :
        print("H0 est rejetée, la condition de normalité des résidus n'est pas vérifiée. Une approche graphique est à réaliser.")
        Validite_list.append({'Condition':'Normalité des résidus','Validité' : 'A vérifier'})
    else: 
        print("H0 est acceptée, la condition de normalité des résidus est vérifiée.")
        Validite_list.append({'Condition':'Normalité des résidus' ,'Validité' : 'Oui'})         
          
          
    # Test 4  
    print(cond4)
    print("Test de Breusch-Pagan :\nH0 : La condition d'homoscédasticité des résidus est vérifiée (p>0.05), les variances des résidus sont constantes \nH1 : La condition d'homoscédasticité des résidus n'est pas vérifiée, les variances des résidus ne sont pas constantes.")      
        # racine carrée des résidus standardisés:
    std_residus = reg.resid/sqrt(sum(reg.resid**2)/(len(reg.resid)-1))
    testBP = sms.het_breuschpagan(np.sqrt(np.abs(std_residus)), reg.model.exog)
    print('La p-value issue du test de Breusch-Pagan est:', testBP[1])       
    if testBP[1] < 0.05:
        print("H0 est rejetée, la condition d'homoscédasticité n'est pas vérifiée. \nLes variances des résidus ne semblent pas constantes. Une approche graphique est à réaliser.")
        Validite_list.append({'Condition':'Homoscédasticité','Validité' : 'A vérifier'})     
    else:
        print("H0 est acceptée, la condition d'homoscédasticité est vérifiée. Les variances des résidus sont constantes.")
        Validite_list.append({'Condition':'Homoscédasticité' ,'Validité' : 'Oui'})
            
    # Test 5
    if nb_Variables > 1:
        print(cond5)
        _, X_vif = dmatrices('margin_low~diagonal+height_left+height_right+margin_up+length', data=dfSansVM, return_type='dataframe')

        dfVif = pd.DataFrame()
        dfVif['variable'] = X_vif.columns
        dfVif['VIF'] = [round(variance_inflation_factor(X_vif.values, i),2) for i in range(X_vif.shape[1])]

        display(dfVif.loc[1:,:])

        if (max((dfVif.loc[1:,:])['VIF'])) > 10 :
            print("\nAu moins une valeur de facteur VIF (Facteur d'Inflation de la Variance) est supérieur à 10. \nIl existe de la collinéarité dans le modèle et la condition 5 n'est pas vérifiée.")
            Validite_list.append({'Condition':'Multicollinéarité' ,'Validité' : 'Non'})
        else : 
            print("\nAucun facteur VIF (Facteur d'Inflation de la Variance) n'est supérieur à 10. \nLa multicollinéarité dans le modèle est faible et la condition 5 est validée.")
            Validite_list.append({'Condition':'Multicollinéarité' ,'Validité' : 'Oui'})  
    else:
        print('')
    
    dfValidite = pd.DataFrame(Validite_list)  
    print("\n\nRésumé des tests de validation des hypothèses du modèle :")
    display(dfValidite)
```

### Coefficients d'un modèle de regression logistique

```python
def do_lreg(df):
    sample = df.sample(df.shape[0], replace=True)
    X_tr = sample[[c for c in sample.columns if c != 'y']]
    y_tr = sample.y

    lr = LogisticRegression(penalty='l2', solver='liblinear')
    lr.fit(X_tr, y_tr)

    params = [lr.intercept_[0]] +  list(lr.coef_[0])

    return params

def Coefficients_stats(X, y):
    lr = LogisticRegression()
    lr.fit(X, y)

    df = pd.DataFrame(X)
    
    df['y'] = y

    r_df = pd.DataFrame([do_lreg(df) for _ in range(100)])

    w = [lr.intercept_[0]] + list(lr.coef_[0])
    se = r_df.std()

    dof = X.shape[0] - X.shape[1] - 1

    summary = pd.DataFrame({
        'Coefficient': w,
        'se': se,
        'z': w / se,
        '.025': w - se,
        '.975': w + se,
        'df': [dof for _ in range(len(w))]
    })
    df = df.drop('y',axis=1)
    summary.index = ["constante"]+list(df.columns)     
    
    
    summary['P>|z|'] = scipy.stats.t.sf(abs(summary.z), df=summary.df)

    return summary.iloc[:,[0,1,2,5,6]]
```

### *Pseudo-R² du modèle*

```python
def efron_rsquare(y, p):
    y_pred = y*1
    n = float(len(y_pred))
    t1 = np.sum(np.power((y_pred - p), 2.0))
    t2 = np.sum(np.power((y_pred- (np.sum((y_pred) / n))), 2.0))
    return round(1.0 - (t1 / t2),2)
```

### Détection des outliers par méthode interquartile

```python
def detect_outliers_IQR(df, col):
    
    #Définir la médiane et la moyenne
    Median = df[col].quantile(0.5).round(2)
    Moyenne = np.around(np.mean(df[col]), decimals=2)
    Ecart = np.around(np.abs(Moyenne-Median), decimals=2)
    print("La médiane est", Median, "et la moyenne est:",Moyenne,". L'écart observé (en valeur absolue) est de",Ecart,".")

    #Définir le 1er quantile ou 25eme percentile
    Q1 = df[col].quantile(0.25)
    print("Le 25eme percentile est", Q1.round(2))
    
    #Définir le 3eme quantile ou 75eme percentile
    Q3 = df[col].quantile(0.75)
    print("Le 75eme percentile est", Q3.round(2))
    
    #Définir l'intervalle InterQuartile
    IQR = Q3-Q1
    print("L'Intervalle InterQuartile est:",IQR.round(2))
  
    # Comptage des outliers
    
    df_outliersIQR = df.loc[(df[col] >= Q3 + 1.5*IQR)|(df[col] <= Q1 - 1.5*IQR),:]
    
    nb_outliersIQR = df_outliersIQR[col].count()      
    
    if nb_outliersIQR > 0:
            
        print("\nAprès analyse de la variable",col,",",nb_outliersIQR,"outliers sont identifiés et présentés ci-dessous.\n")

        df_outliersIQR = df_outliersIQR.sort_values(col)
        display(df_outliersIQR)
       
        
    else:
        print("\n\nAprès analyse de la variable",col,", aucun outlier n'a été détecté.\n")
       
```

#### *Modèles de régression linéaire et indicateurs qualité pour le choix de la variable*

```python
# régressions à une variable

dfModelQuality = pd.DataFrame()
a_list = []
b_list = []
R2_list = []
RMSE_list = []
               
               
for v in V_RLS_liste:
    X = dfSansVM[[v]]
    y = dfSansVM['margin_low']
    
# Calcul de l'équation obtenue par le modèle
    model = linear_model.LinearRegression()
    results = model.fit(X, y)
    
    # coefficients a et b
    a = results.coef_[0]
    b = results.intercept_
    a_list.append(round(a,5))
    b_list.append(round(b,5))
    #print("\n\n__________________________________\n\nLe modèle de régression linéaire simple de type y=ax+b entre les variables margin_low (y) et",v, "indiquerait que \na=", a,"\nb=",b)
    
    # indicateur r2
    y_pred = (results.coef_[0])*(dfSansVM[v])+(results.intercept_)
    r2 = r2_score(y,y_pred)
    #print("R-Squared = %.3f" % r2)
    R2_list.append(round(r2,3))
    
    # indicateur Root Square Mean Error
    mean_y = y.mean()
    RMSE = mean_squared_error(y,y_pred, squared = False)
    RMSEm = round((100*RMSE/(mean_y)),1)
    #print("RMSE: %.2f" % RMSE,"(",RMSEm,"%)")
    RMSE_list.append(RMSEm)
    
dfModelQuality['Variable explicative'] = V_RLS_liste
dfModelQuality['a'] = a_list
dfModelQuality['b'] = b_list
dfModelQuality['R²'] = R2_list
dfModelQuality['RMSE (%)'] = RMSE_list

print('\nParamètres des modèles linéaires testés pour évaluer margin_low :')
dfModelQuality
```

#### *Significativité du modèle de régression linéaire multiple* 

Considérons l'équation y = Σ(*ai*.xi)+*b* où y est la variable *margin_low* et x les variables explicatives,  

Soit H0 l'hypothèse nulle sur le modèle global : Tous les coefficients *ai* et *b* de la régression sont nuls  
Et H1 l'hypothèse alternative : Au moins un des coefficients n'est pas nul

H0 est rejetée suite au test de Fisher (p(F-statistic) < 0.05), ce qui veut dire qu'au moins un des coefficients est non nul. 

<span style = 'color:darkgreen; font-family:Microsoft JhengHei UI Light'> <b>
► Le modèle est globalement significatif.  

La démarche pas-à-pas nous amène à optimiser ce modèle par un test de student sur chaque coefficient. Dans cette approche backward, on commence avec toutes les variables puis on ote progressivement les variables ayant le coefficent non significativement différents de 0 (On ote la variable qui a la plus forte p-value à chaque itération, tant qu'il y a une p-value>0.05).

<span style = 'color:darkgreen; font-family:Microsoft JhengHei UI Light'> <b>
► Le modèle optimisé se base sur la prédiction de margin_low par les variables *is_genuine* et *margin_up*.

- Le nombre de faux positifs FP  
- Le nombre de vrais négatifs VN  
- Le nombre de faux négatifs FN  

Pour calculer les indicateurs  : 

Accuracy : L'exactitude est la proportion de prédictions correctes parmi le total des prédictions. C'est un bon indicateur lorsque les classes sont équilibrées.

$$\large Accuracy = \dfrac{TP + TN}{TP+TN+FP+FN}$$



Precision : La précision est la proportion de prédictions positives qui sont réellement positives.   

$$\large Precision = \dfrac{TP }{TP+FP}$$

Recall (ou sensibilité) : Le rappel est la proportion de vrais positifs qui sont correctement prédits comme tels.   

$$\large Recall = \dfrac{TP }{TP+FN}$$

F1-score : Le score F1 est une moyenne harmonique de la précision et du rappel. Il donne la contribution de ces deux mesures.   
Plus le score F1 est élevé et plus le modèle est performant (spécifique et sensible).   

$$\large F1 = 2\times \dfrac{Precision\times Recall }{Precision+Recall}$$

Le F1-score est un indicateur particulièrement utilisé pour les problèmes utilisant des données déséquilibrées comme la détection de fraudes ou la prédiction d’incidents graves. Ce métrique sera donc plus adapté à notre cas.

## <a name="A211"><span style='color:#635c5c'> II.1.1. Dataframe de travail pour méthode non supervisée </span></a>

## <a name="A212"><span style='color:#635c5c'> II.1.2. Détermination du nombre de clusters </span></a>  

```python
from sklearn.cluster import KMeans

K = df_non_supervised.values

# liste vide pour enregistrer les inerties :  
intertia_list = [ ]

# liste de nombres de clusters : 
k_list = range(1, 6)

# Pour chaque nombre de clusters : 
for k in k_list : 
    
    # k-means pour k clusters
    kmeans = KMeans(n_clusters=k,n_init=1, random_state = 1000)
    
    # entrainement
    kmeans.fit(K)
    
    # enregistrer l'inertie obtenue : 
    intertia_list.append((kmeans.inertia_)/10)

fig, ax = plt.subplots(1,1,figsize=(8,4))

ax.set_ylabel("intertie")
ax.set_xlabel("nombre de clusters")
ax.set_title("Détermination du nombre de clusters par la méthode du coude.")

ax = plt.plot(k_list, intertia_list, marker='+')
```

<span style = 'color:darkgreen; font-family:Microsoft JhengHei UI Light'> <b>
► Notre objectif de classer en deux groupes est réalisable directement par le K-means (sans ACP préalable).

Le modèle vise à classer les billets en catégorie "vrai" ou "faux". Dans la régression, la variable y sera donc booléenne et les variables x seront continues. La régression logistique est donc choisie pour mettre en relation ces deux types de variables.

La courbe ROC (Receiver Operating Characteristic) est utilisée pour évaluer la performance du modèle en traçant le taux de vrais positifs (sensibilité, 99.5%) en fonction du taux de faux positifs (1-spécificité, 4.2%).

L'AUC (Area Under the Curve) est l'aire sous la courbe ROC. Elle fournit une mesure agrégée de la performance du modèle de classification à tous les seuils de classification possibles. Elle peut aussi être interprétée comme la probabilité qu'un classificateur rangera un exemple positif aléatoirement choisi plus haut qu'un exemple négatif aléatoirement choisi. Les valeurs vont de 0 à 1, où une valeur de 0,5 correspond à un modèle aléatoire et une valeur de 1 correspond à un modèle parfait.

ROC-AUC score est la valeur numérique pour l'AUC. Il est utilisé comme une mesure de la performance globale du modèle de classification.

```python
def detecteur_faux_billets(csv_file):
    """"
    Données d'entrées 
        - un fichier CSV standardisé
    

    Paramètres :
    csv_file (str) : Le chemin vers le fichier CSV
                     et des colonnes pour chaque variable indépendante.
    model : Un modèle de régression logistique entraîné.

    Retourne un DataFrame contenant 
    - les données originales du fichier csv, 
    - la prédiction attestant l'authenticité du billet ou non 
    - la fiabilité de la prédiction pour chaque billet (probabilité de se tromper) et son indicateur associé
                     
    """
    

    # Charger le fichier CSV
    df_billets = pd.read_csv(csv_file)
    data = df_billets.iloc[:,[0,1,2,3,4,5]].to_numpy()
    #data = np.around(data,2)
    df_billets_new = pd.DataFrame(data=data, columns=(df_billets.iloc[:,[0,1,2,3,4,5]]).columns)
    
    # Charger le modèle et l'appliquer
    joblib.dump(reg_logit,'regression_logistique')
    RegLog_model = joblib.load('regression_logistique')
    Authentification = RegLog_model.predict(df_billets_new)
    
    #Transformer la colonne résultat boléen en format string
    
    Authenticite = ['Billet authentique' if pred == 1 else 'Faux billet' for pred in Authentification]
    
    # Créer le tableau résultat
    df_resultat = pd.DataFrame()
        
        # Définir l'authenticité du billet
    df_resultat['Authenticité'] = Authenticite
    
        # Définir les billets par un identifiant
    #id_list = df_resultat.index.to_list()
    df_resultat['ID'] = df_billets.iloc[:,6]
    df_resultat = df_resultat.reindex(columns=['ID','Authenticité'])
    
        # Calculer les probabilités associées au classement
    proba = RegLog_model.predict_proba(df_billets_new)
    
    proba_list=[]
    
    for i in range (0,len(df_resultat['Authenticité'])):
        if df_resultat.iloc[i,1]=='Billet authentique':
            p = proba[i, 0]
            proba_list.append(p*100)
        else:
            p = proba[i, 1]
            proba_list.append((p*100))
           
    df_resultat["Probabilité de se tromper (%)"] = proba_list
    
    Indicateur_fiabilite_list = []
  
    
    for i in range (0,len(df_resultat['Authenticité'])):
        
        
        if df_resultat.iloc[i,2] < 1 :
            F = 'Bonne'
            Indicateur_fiabilite_list.append(F)
            
        elif df_resultat.iloc[i,2] < 5 :
            F = 'Moyenne'
            Indicateur_fiabilite_list.append(F)
                       
        else : 
            F = 'INSUFFISANTE'
            Indicateur_fiabilite_list.append(F)

    
    df_resultat["Fiabilité de la prédiction"] = Indicateur_fiabilite_list
    
    # Ajouter les caractéristiques des billets au DataFrame
    df_detection = pd.concat([df_resultat,df_billets_new], axis=1)
    

    def Alerte_fiabilite(s, seuil_alerte, column):
        is_max = pd.Series(data=False, index=s.index)
        is_max[column] = s.loc[column] >= seuil_alerte
        return ['color: red' if is_max.any() else '' for v in is_max] 
    
    # Liste des billets à vérifier sur d'autres critères:
    ID_list=[]
    for l in range (0,len(df_detection)):
        if df_detection.iloc[l,2] < 5 :
            continue
        else :
            id = df_detection.iloc[l,0]
            ID_list.append(id)
            
    print("Nombre de billets à vérifier de manière plus approfondie :", len(ID_list))
    if (len(ID_list)) > 0 :
        print("Il s'agit de : ",(" , ".join(ID_list)))
         

    else:   
        print("")
    return df_detection.style.apply(Alerte_fiabilite, seuil_alerte = 5, column=['Probabilité de se tromper (%)'], axis=1).format(precision=2)
    
```
