### <a href="#RM1" name="M1" style='text-decoration:none'><span style='color:black'>*Test d'unicité*</span></a>

```python
# Effectuer un test d'unicité sur colonne clé candidate "cle" en vue de valider son utilisation en tant que clé primaire.
    
def test_unicite(dataframe, cle):
    
    doublons = dataframe[cle].duplicated().sum()
    
        # Afficher le résultat du test de manière explicite.

    if doublons > 0 :
            print("La colonne contient",doublons,"doublons. \nle critère d'unicité, obligatoire pour une clé primaire, n'est pas respecté.")          
           
    else:
            print("Le critère d'unicité est respecté. La clé primaire ne contient pas de doublons.")
          
```

### *Test d'intégrité référentielle d'une clé étrangère*

```python
def test_cle_etrangere(df_primaire, df_etrangere, cle):
    
    liste_cle_primaire = df_primaire[cle].to_list()
    
    df_error = df_etrangere.loc[~df_etrangere[cle].isin(liste_cle_primaire),:]
    error = df_error[cle].drop_duplicates(keep='first')
    
    
    if len(error)==0:
        print("La colonne",cle,"respecte la contrainte d'intégrité référentielle d'une clé étrangère. \nToutes les valeurs de la clé étrangère",cle,"sont référencées dans la table d'origine.")
    
    elif len(error)==1:
        for e in error:
            print("La colonne",cle,"ne respecte pas la contrainte d'intégrité référentielle d'une clé étrangère.\nEn effet, la valeur",e,"n'est pas référencée dans la table d'origine.")
    
    else:
        print("La colonne",cle,"n'est pas conforme à la contrainte de clé étrangère.\nEn effet, les valeurs ci-dessous ne sont pas référencée dans la table d'origine:")
        for e in error:
            print(e)
```

### <a href="#RM2" name="M2" style='text-decoration:none'><span style='color:black'>*Détection des outliers par la méthode interquartile*</span></a>

```python
def detect_outliers_IQR(df, col):
    
    #Définir la médiane et la moyenne
    Median = df[col].quantile(0.5).round(2)
    Moyenne = np.around(np.mean(df[col]), decimals=2)
    Ecart = np.around(np.abs(Moyenne-Median), decimals=2)
    print("La médiane est", Median, "et la moyenne est:",Moyenne,". L'écart observé (en valeur absolue) est de",Ecart,".")

    #Définir le 1er quantile ou 25eme percentile
    Q1 = df[col].quantile(0.25)
    print("Le 25eme percentile est", Q1.round(2))
    
    #Définir le 3eme quantile ou 75eme percentile
    Q3 = df[col].quantile(0.75)
    print("Le 75eme percentile est", Q3.round(2))
    
    #Définir l'intervalle InterQuartile
    IQR = Q3-Q1
    print("L'Intervalle InterQuartile est:",IQR.round(2))
  
    # Comptage des outliers
    
    df_outliersIQR = df.loc[(df[col] >= Q3 + 1.5*IQR)|(df[col] <= Q1 - 1.5*IQR),:]
    
    nb_outliersIQR = df_outliersIQR[col].count()      
    
    if nb_outliersIQR > 0:
            
        print("\nAprès analyse de la variable",col,",",nb_outliersIQR,"outliers sont identifiés et présentés ci-dessous.\n")

        df_outliersIQR = df_outliersIQR.sort_values(col)
        display(df_outliersIQR)
       
        
    else:
        print("\n\nAprès analyse de la variable",col,", aucun outlier n'a été détecté.\n")
       
```

## <a name="A24"> <span style='color:#aa2d22'> II. 4. Répartition du chiffre d'affaires entre nos clients</span></a>

### *Courbe de Lorenz*

```python
# créer le tableau du chiffre d'affaire par client pour faire la courbe de lorenz
dfCA_client_lorenz = dfAnalyse.groupby(['client_id']).sum()
dfCA_client_lorenz = dfCA_client_lorenz.drop(['birth','categ','age'], axis=1).reset_index()

# Créer la courbe de Lorenz
CA_lorenz = dfCA_client_lorenz['price'].values
n = len(CA_lorenz)
print("La courbe de Lorenz du chiffre d'affaires des 2 ans d'exercice, pour nos",n,"clients est:")
lorenz = np.cumsum(np.sort(CA_lorenz)) / CA_lorenz.sum()
lorenz = np.append([0],lorenz) 

xaxis = np.linspace(0-1/n,1+1/n,len(lorenz)) 
plt.title("Courbe de Lorenz du chiffre d'affaires sur 2 ans par client", y=1.05,fontsize=16)
plt.ylabel("Proportion du chiffre d'affaires",  fontsize=14)
plt.xlabel("Clients", fontsize=14)
plt.plot(xaxis,lorenz,drawstyle='steps-post')

#tracer la bisséctrice
plt.plot([0,1], [0,1]) 
plt.show()
```

<span style = 'color:darkgreen; font-family:Microsoft JhengHei UI Light'> <b> ► Près de la moitié du chiffre d'affaires est généré par près de 20% des clients.  
► La pointe de la courbe montre un décroché pour les clients les plus importants.  
► Cela suppose un intérêt pour la catégorisation de certains clients en "clients grand-compte". </b> </span> 

### *Indice de Gini*

```python
#Calculer l'indice de Gini
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = 2*S
gini = gini.round(2)
### *Courbe de Lorenz
print("L'indice de Gini est:",gini)
```

<span style = 'color:darkgreen; font-family:Microsoft JhengHei UI Light'> <b> ► L'indice de Gini montre des inégalités entre les profils d'achat des clients.   
 ► En d'autres termes, certains clients dépensent beaucoup et d'autres très peu.  
 ► Cependant, l'indice étant à proche de 0.5, cela indique également que nous ne sommes pas dans un cas extrême où nous aurions seulement deux profils "gros clients *vs.* petits clients", mais que nous sommes en présence d'une large gamme de profils clients intermédiaires. </b> </span>
